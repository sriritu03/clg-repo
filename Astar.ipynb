```python
#BFS
from collections import deque
graph = {
    0: [(1,2), (2,1)],
    1: [(0, 2),(2, 5),(4,3),(3,11)],
    2: [(0,1),(1, 5),(4,1),(5,15)],
    3: [(1,11), (4,2),(6,1)],
    4: [(1,3), (2, 1),(3,2),(6,5),(5,4)],
    5: [(4, 4), (2, 15)],
    6: [(3,1),(4,5),(5,1)]
}
heuristic = {
    0: 1,
    1: 3,
    2: 15,
    3: 2,
    4: 1,
    5: -1,
    6: 0
}
def bfs(graph, start, goal):
    queue = deque([(start, [start], 0)])  # (node, path, cost)
    visited = set()

    while queue:
        node, path, cost = queue.popleft()

        if node == goal:
            return path, cost

        if node not in visited:
            visited.add(node)

            for neighbor, weight in graph.get(node,[]):
                queue.append((neighbor, path + [neighbor], cost + weight))

    return None, 0
start_node = 0
goal_node = 6

print("BFS:")
path, cost = bfs(graph, start_node, goal_node)
print(f"Path: {path}, Cost: {cost}\n")
```

```python
import heapq
graph = {
    0: [(1,2), (2,1)],
    1: [(0, 2),(2, 5),(4,3),(3,11)],
    2: [(0,1),(1, 5),(4,1),(5,15)],
    3: [(1,11), (4,2),(6,1)],
    4: [(1,3), (2, 1),(3,2),(6,5),(5,4)],
    5: [(4, 4), (2, 15)],
    6: [(3,1),(4,5),(5,1)]
}
heuristic = {
    0: 1,
    1: 3,
    2: 15,
    3: 2,
    4: 1,
    5: -1,
    6: 0
}
def ucs(graph, start, goal):
    pq = []
    heapq.heappush(pq, (0, start, [start]))  # (cost, node, path)

    while pq:
            cost, node, path = heapq.heappop(pq)

        if node == goal:
            return path, cost

        for neighbor, weight in graph[node]:
            heapq.heappush(pq, (cost + weight, neighbor, path + [neighbor]))

    return None, 0
print("UCS:")
start_node = 0
goal_node = 6
path, cost = ucs(graph, start_node, goal_node)
print(f"Path: {path}, Cost: {cost}\n")
```

```python
import heapq
graph = {
    0: [(1,2), (2,1)],
    1: [(0, 2),(2, 5),(4,3),(3,11)],
    2: [(0,1),(1, 5),(4,1),(5,15)],
    3: [(1,11), (4,2),(6,1)],
    4: [(1,3), (2, 1),(3,2),(6,5),(5,4)],
    5: [(4, 4), (2, 15)],
    6: [(3,1),(4,5),(5,1)]
}
heuristic = {
    0: 1,
    1: 3,
    2: 15,
    3: 2,
    4: 1,
    5: -1,
    6: 0
}
def Astar(graph,goal,start,heuristic):
    pq=[]
    heapq.heappush(pq,(heuristic[start],0,[start],start))
    visited=set()
    while pq:
        curr_heuristic,curr_cost,path,node=heapq.heappop(pq)
        if node == goal:
            return (path,curr_cost)
        if node in visited:
            continue
        visited.add(node)
        for neighbour,weight in graph.get(node,[]):
            if neighbour not in visited:
                new_cost=curr_cost+weight;
                new_heuristic=new_cost+heuristic[neighbour]
                heapq.heappush(pq,(new_heuristic,new_cost,[neighbour]+path,neighbour))
    return (0,0)

print("A* Search:")
start_node = 0
goal_node = 6
path, cost = Astar(graph, start_node, goal_node, heuristic)
print(f"Path: {path}, Cost: {cost}")

```
